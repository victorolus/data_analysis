```python
# import the packages to be used in this project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv(r'/Users/victorolushola/Documents/movies.csv')
```

```python
# looking at the data

df
```

```python
# checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#dropping the empty rows

df = df.dropna()
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
#change data type of columns


df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Correlation Matrix between all numeric columns


non_numeric_columns = df.select_dtypes(exclude=['float64', 'int64']).columns
print("Non-numeric columns:", non_numeric_columns)
```

```python
df_numeric = df.drop(non_numeric_columns, axis=1)
```

```python
#df_numeric = df_numeric.apply(pd.to_numeric, errors='coerce')
df_numeric.corr(method='pearson')
```

```python
df_numeric.corr(method ='kendall')
```

```python
df_numeric.corr(method ='spearman')
```

```python
correlation_matrix = df_numeric.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Assuming df_numeric is the numeric DataFrame after handling non-numeric values
correlation_matrix = df_numeric.corr(method='pearson')

# Flatten the correlation matrix into a list of correlation pairs with column names
correlation_list = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i + 1, len(correlation_matrix.columns)):
        col1, col2 = correlation_matrix.columns[i], correlation_matrix.columns[j]
        correlation_value = correlation_matrix.iloc[i, j]
        correlation_list.append((col1, col2, correlation_value))

# Sort the list based on the absolute correlation values
sorted_correlation_list = sorted(correlation_list, key=lambda x: abs(x[2]))

# Print the table with the weakest to strongest correlation pairs
print("Weak to Strong Correlation Pairs:")
print("{:<15} {:<15} {:<15}".format("Column 1", "Column 2", "Correlation"))
print("-" * 45)
for pair in sorted_correlation_list:
    print("{:<15} {:<15} {:<15}".format(pair[0], pair[1], pair[2]))
```

```python
# Group by company and calculate the mean revenue for each
revenue_by_company = df.groupby('company')['gross'].mean().sort_values(ascending=False)

# Plot the top N companies by mean revenue
top_n = 10  # Adjust this value based on how many top companies you want to visualize
plt.figure(figsize=(12, 6))
sns.barplot(x=revenue_by_company.head(top_n).index, y=revenue_by_company.head(top_n))
plt.title(f'Top {top_n} Companies by Mean Revenue')
plt.xlabel('Company')
plt.ylabel('Mean Revenue')
plt.xticks(rotation=90)
plt.show()
```
